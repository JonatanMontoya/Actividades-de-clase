Ejercico 10

Generar muestras e histograma de las funciones: triangular(), betavariate(), expovariate(), gammavariate(), lognormvariate(), normalvariate(), vonmisesvariate(), paretovariate() y weibullvariate()

```python
#Montoya López Jonatan Emmanuel
import random
import numpy as np
import matplotlib.pyplot as plt

nums = []
mu = 100
sigma = 50

for i in range(200):
    temp = random.gauss(mu, sigma)
    nums.append(temp)
plt.plot(nums, color = 'grey')
plt.show()

for i in range(10000):
    temp = random.gauss(mu, sigma)
    nums.append(temp)
plt.hist(nums, bins = 200, color = 'grey')
plt.show()
```
