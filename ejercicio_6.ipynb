Ejercicio 6:

Repetir el experimento, utilizando un cuarto de elipse

```python
#Montoya Lpez Jonatan Emmanuel
import random
import matplotlib.pyplot as plt
N = 1000
def pi_Taylor(n):
  L = []
  m = 0
  for i in range(n):
    x_1 = random.random()
    y_1 = random.random()
    z = ((x_1**2)+(y_1**2))**(1/2)
    if z < 1:
        m = m+1
    pi = 4*(m/(i+1))
    print(i,pi)
    L.append(pi) 
  plt.plot(L, color = "blue", markersize=1)
  plt.xlabel("Iteraciones")
  plt.ylabel("Estimacin")
  plt.grid()
  plt.show()
pi_Taylor(N)
```
