Ejercicio 3:

Graficar las salidas obtenidas en la Tarea 1

```python
from random import random
import matplotlib.pyplot as plt
N=10000000
n=0
lx=[]
ly=[]
for i in range (N):
  x=random()
  y=random()
  if x**2+y**2<1:
    n=n+1
  i=i+1
  if i%10000==0:
    est=4*n/(i)
    lx.append(i)
    ly.append(est)
plt.plot(lx,ly)
plt.show()
```
